# ***Variation Inflation Factor***

A set of multiple regression variables' variance inflation factor (VIF) is a gauge of how multicollinear they are. The VIF for a regression model variable is mathematically equivalent to the ratio of the variance of the entire model to the variance of a model with only that one independent variable. Calculated for each independent variable is this ratio. A high VIF denotes a high degree of collinearity between the associated independent variable and the other variables in the model.

#### Chart - 14 - Correlation Heatmap with Variance Inflation Factor (VIF) on.

````python
# Multicorrelinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calculate_vif(X):

  # calculating VIF
  vif =pd.DataFrame()
  vif["variables"] = X.columns
  vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X. shape[1])]

  return(vif)

calculate_vif(df[[i for i in df.describe().columns if i not in ['Date', 'Close']]])
     
````

Despite our strong VIF scores, we will not perform feature engineering because each feature is critical for this particular use case. Most indicators in the real world consider each of these characteristics to forecast future values.

We must exclude the features from our dataset that have the lowest correlation to the dependent variable since they have a very high VIF of Open, High, and Low features.


````python
# Here we will drop the open feature from the VIF matrix
calculate_vif(df[[i for i in df.describe().columns if i not in ['Date', 'Close','Open']]])
````

### Metrics Comparison




````python
# Function to run all model 
def calculate_score_model(X_train,y_train,X_test,y_test):
    df_columns=[]
    df=pd.DataFrame(columns=df_columns) # Creating dataframe to store the train and test metrics for each of the models
    
    i=0
    
    #Reading model one by one
    for model in models:
        model.fit(X_train,y_train)
        
        y_pred_train = model.predict(X_train)
        y_pred_test = model.predict(X_test)

        #Computing evaluation metrics
        train_accuracy=model.score(X_train,y_train)
        test_accuracy=model.score(X_test,y_test)
        
        MAE=metrics.mean_absolute_error(y_test, y_pred_test)
        MSE=metrics.mean_squared_error(y_test, y_pred_test)
        RMSE=math.sqrt(MSE)
        
        Rsquare=metrics.r2_score(y_test, y_pred_test)
        
        
        #Inserting in dataframe
        df.loc[i,"Model_Name"]=model.__class__.__name__
        df.loc[i,"MAE"]=round(MAE,3)
        df.loc[i,"MSE"]=round(MSE,3)
        df.loc[i,"RMSE"]=round(RMSE,3)
        # df.loc[i,"MAPE"]=round(MAPE,3)
        df.loc[i,"Rsquare"]=round(Rsquare,3)
        
        i+=1
    
    #Sorting values by accuracy
    df.sort_values(by=['Rsquare'],ascending=False,inplace=True)
    
    return df
````
